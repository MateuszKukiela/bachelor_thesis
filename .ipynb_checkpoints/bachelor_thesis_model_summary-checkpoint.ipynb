# Imports

```python
import mne
import numpy as np
import os
import braindecode
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_path = './models/'
data_type = 'XDAWN'
pickle_in = open(model_path + f'ShallowFBCSPNet_{data_type}.pickle', "rb")
model_XDAWN = pickle.load(pickle_in)
```

```python
data_type = 'averaged_2'
pickle_in = open(model_path + f'ShallowFBCSPNet_{data_type}.pickle', "rb")
model_averaged_2 = pickle.load(pickle_in)
```

```python
data_type = 'averaged_3'
pickle_in = open(model_path + f'ShallowFBCSPNet_{data_type}.pickle', "rb")
model_averaged_3 = pickle.load(pickle_in)
```

```python
data_type = 'averaged_5'
pickle_in = open(model_path + f'ShallowFBCSPNet_{data_type}.pickle', "rb")
model_averaged_5 = pickle.load(pickle_in)
```

```python
data_type = 'averaged_10'
pickle_in = open(model_path + f'ShallowFBCSPNet_{data_type}.pickle', "rb")
model_averaged_10 = pickle.load(pickle_in)
```

```python
data_type = 'clean'
pickle_in = open(model_path + f'ShallowFBCSPNet_{data_type}.pickle', "rb")
model_clean = pickle.load(pickle_in)
```

```python
plt.figure(figsize=(16,12))
cell = 'valid_misclass'
model_XDAWN.epochs_df[cell].plot(label='XDAWN valid_misclass')
model_averaged_2.epochs_df[cell].plot(label=f'averaged_2 {cell}')
model_averaged_3.epochs_df[cell].plot(label=f'averaged_3 {cell}')
model_averaged_5.epochs_df[cell].plot(label=f'averaged_5 {cell}')
model_averaged_10.epochs_df[cell].plot(label=f'averaged_10 {cell}')
model_clean.epochs_df[cell].plot(label=f'clean {cell}')

plt.legend()
plt.show()
```
