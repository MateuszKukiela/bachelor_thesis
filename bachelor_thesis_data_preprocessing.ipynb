# Imports

```python
import mne
import numpy as np
import os
```

# EEGLAB data import with ICA

```python
point = 200
X_pseudo_test = np.zeros((0, 19, 306))
X_pseudo_train = np.zeros((0, 19, 306))
X_word_test = np.zeros((0, 19, 306))
X_word_train = np.zeros((0, 19, 306))
path = './data'
for filename in os.listdir(path):
    filename = path + '/' + filename 
    if '.set' in filename:
        print(filename)
        raw = mne.io.read_epochs_eeglab(filename)
        raw_train = raw[:point]
        raw_test = raw[point:]
        
        # Train
        ica = mne.preprocessing.ICA(n_components=10, random_state=97, max_iter=800)
        ica.fit(raw_train)
        ica.detect_artifacts(raw_train)
        ica.apply(raw_train)
        raw_train.apply_baseline(baseline=(None,0))
        raw_train.filter(l_freq=1, h_freq=40)
        sig_train = raw_train.get_data()
        
        #Test
        ica = mne.preprocessing.ICA(n_components=10, random_state=97, max_iter=800)
        ica.fit(raw_test)
        ica.detect_artifacts(raw_test)
        ica.apply(raw_test)
        raw_test.apply_baseline(baseline=(None,0))
        raw_test.filter(l_freq=1, h_freq=40)
        sig_test = raw_test.get_data()
        
        if '_exp1.set' in filename:
            X_word_train = np.concatenate((X_word_train, sig_train), axis = 0)
            X_word_test = np.concatenate((X_word_test, sig_test), axis = 0)
        if '_exp2.set' in filename:
            X_pseudo_train = np.concatenate((X_pseudo_train, sig_train), axis = 0)
            X_pseudo_test = np.concatenate((X_pseudo_test, sig_test), axis = 0)
```

```python
output = './processed_data/'
data_type = 'ICA'
np.save(output + f'X_pseudo_test_{data_type}.npy', X_pseudo_test)
np.save(output + f'X_word_test_{data_type}.npy', X_word_test)
np.save(output + f'X_pseudo_train_{data_type}.npy', X_pseudo_train)
np.save(output + f'X_word_train_{data_type}.npy', X_word_train)
```

# EEGLAB data import with XDAWN denoising

```python
point = 200
X_pseudo_test = np.zeros((0, 19, 306))
X_pseudo_train = np.zeros((0, 19, 306))
X_word_test = np.zeros((0, 19, 306))
X_word_train = np.zeros((0, 19, 306))
path = './data'
for filename in os.listdir(path):
    filename = path + '/' + filename 
    if '.set' in filename:
        print(filename)
        raw = mne.io.read_epochs_eeglab(filename)
        raw.apply_baseline(baseline=(None,0))
        raw.filter(l_freq=1, h_freq=40)
        raw_train = raw[:point]
        raw_test = raw[point:]
        
        #XDAWN denosing test
        temp_syg = raw_test.get_data()
        temp_syg = temp_syg.reshape([temp_syg.shape[1],temp_syg.shape[0]*temp_syg.shape[2]])
        signal_cov = mne.compute_raw_covariance(mne.io.RawArray(temp_syg, raw_test[0].info))
        xd = mne.preprocessing.Xdawn(n_components=2, signal_cov=signal_cov)
        xd.fit(raw_test)
        epochs_denoised = xd.apply(raw_test)
        raw_test = next(iter(epochs_denoised.values()))
        sig_test = raw_test.get_data()
        
        #XDAWN denosing train
        temp_syg = raw_train.get_data()
        temp_syg = temp_syg.reshape([temp_syg.shape[1],temp_syg.shape[0]*temp_syg.shape[2]])
        signal_cov = mne.compute_raw_covariance(mne.io.RawArray(temp_syg, raw_train[0].info))
        xd = mne.preprocessing.Xdawn(n_components=2, signal_cov=signal_cov)
        xd.fit(raw_train)
        epochs_denoised = xd.apply(raw_train)
        raw_train = next(iter(epochs_denoised.values()))
        sig_train = raw_train.get_data()
        
        if '_exp1.set' in filename:
            X_word_train = np.concatenate((X_word_train, sig_train), axis = 0)
            X_word_test = np.concatenate((X_word_test, sig_test), axis = 0)
        if '_exp2.set' in filename:
            X_pseudo_train = np.concatenate((X_pseudo_train, sig_train), axis = 0)
            X_pseudo_test = np.concatenate((X_pseudo_test, sig_test), axis = 0)
```

```python
output = './processed_data/'
data_type = 'XDAWN'
np.save(output + f'X_pseudo_test_{data_type}.npy', X_pseudo_test)
np.save(output + f'X_word_test_{data_type}.npy', X_word_test)
np.save(output + f'X_pseudo_train_{data_type}.npy', X_pseudo_train)
np.save(output + f'X_word_train_{data_type}.npy', X_word_train)
```

# EEGLAB data import averaged

```python
random_size = 10
point = 200
X_pseudo_test = np.zeros((0, 19, 306))
X_pseudo_train = np.zeros((0, 19, 306))
X_word_test = np.zeros((0, 19, 306))
X_word_train = np.zeros((0, 19, 306))
path = './data'
for filename in os.listdir(path):
    filename = path + '/' + filename 
    if '.set' in filename:
        print(filename)
        raw = mne.io.read_epochs_eeglab(filename)
        raw.apply_baseline(baseline=(None,0))
        raw.filter(l_freq=1, h_freq=40)
        sig = raw.get_data()
        
        # Mean train signal
        sig_train_mean = np.zeros(sig[:point,:,:].shape)
        for i in range(sig[:point,:,:].shape[0]):
            x = np.random.randint(sig[:point,:,:].shape[0], size=random_size)
            sig_train_mean[i,:,:] = np.mean(sig[:point,:,:][x,:,:], axis=0)
        
        # Mean test signal
        sig_test_mean = np.zeros(sig[point:,:,:].shape)
        for i in range(sig[point:,:,:].shape[0]):
            x = np.random.randint(sig[point:,:,:].shape[0], size=random_size)
            sig_test_mean[i,:,:] = np.mean(sig[point:,:,:][x,:,:], axis=0)
        
        if '_exp1.set' in filename:
            X_word_train = np.concatenate((X_word_train, sig_train_mean), axis = 0)
            X_word_test = np.concatenate((X_word_test, sig_test_mean), axis = 0)
        if '_exp2.set' in filename:
            X_pseudo_train = np.concatenate((X_pseudo_train, sig_train_mean), axis = 0)
            X_pseudo_test = np.concatenate((X_pseudo_test, sig_test_mean), axis = 0)
```

```python
output = './processed_data/'
data_type = 'averaged'
np.save(output + f'X_pseudo_test_{data_type}_{random_size}.npy', X_pseudo_test)
np.save(output + f'X_word_test_{data_type}_{random_size}.npy', X_word_test)
np.save(output + f'X_pseudo_train_{data_type}_{random_size}.npy', X_pseudo_train)
np.save(output + f'X_word_train_{data_type}_{random_size}.npy', X_word_train)
```

# EEGLAB data import with XDAWN denoising and validation set

```python
point = 200
X_pseudo_test = np.zeros((0, 19, 306))
X_pseudo_train = np.zeros((0, 19, 306))
X_word_test = np.zeros((0, 19, 306))
X_word_train = np.zeros((0, 19, 306))

X_word_valid = np.zeros((0, 19, 306))
X_pseudo_valid = np.zeros((0, 19, 306))

path = './data'
for i, filename in enumerate(os.listdir(path)):
    filename = path + '/' + filename 
    if '.set' in filename:
        print(filename)
        raw = mne.io.read_epochs_eeglab(filename)
        raw.apply_baseline(baseline=(None,0))
        raw.filter(l_freq=1, h_freq=40)
        if i < 18:
            temp_syg = raw.get_data()
            temp_syg = temp_syg.reshape([temp_syg.shape[1],temp_syg.shape[0]*temp_syg.shape[2]])
            signal_cov = mne.compute_raw_covariance(mne.io.RawArray(temp_syg, raw[0].info))
            xd = mne.preprocessing.Xdawn(n_components=2, signal_cov=signal_cov)
            xd.fit(raw)
            epochs_denoised = xd.apply(raw)
            raw = next(iter(epochs_denoised.values()))
            sig = raw.get_data()
            
            if '_exp1.set' in filename:
                X_word_valid = np.concatenate((X_word_valid, sig), axis = 0)
            if '_exp2.set' in filename:
                X_pseudo_valid = np.concatenate((X_pseudo_valid, sig), axis = 0)
        else:
            raw_train = raw[:point]
            raw_test = raw[point:]

            #XDAWN denosing test
            temp_syg = raw_test.get_data()
            temp_syg = temp_syg.reshape([temp_syg.shape[1],temp_syg.shape[0]*temp_syg.shape[2]])
            signal_cov = mne.compute_raw_covariance(mne.io.RawArray(temp_syg, raw_test[0].info))
            xd = mne.preprocessing.Xdawn(n_components=2, signal_cov=signal_cov)
            xd.fit(raw_test)
            epochs_denoised = xd.apply(raw_test)
            raw_test = next(iter(epochs_denoised.values()))
            sig_test = raw_test.get_data()

            #XDAWN denosing train
            temp_syg = raw_train.get_data()
            temp_syg = temp_syg.reshape([temp_syg.shape[1],temp_syg.shape[0]*temp_syg.shape[2]])
            signal_cov = mne.compute_raw_covariance(mne.io.RawArray(temp_syg, raw_train[0].info))
            xd = mne.preprocessing.Xdawn(n_components=2, signal_cov=signal_cov)
            xd.fit(raw_train)
            epochs_denoised = xd.apply(raw_train)
            raw_train = next(iter(epochs_denoised.values()))
            sig_train = raw_train.get_data()

            if '_exp1.set' in filename:
                X_word_train = np.concatenate((X_word_train, sig_train), axis = 0)
                X_word_test = np.concatenate((X_word_test, sig_test), axis = 0)
            if '_exp2.set' in filename:
                X_pseudo_train = np.concatenate((X_pseudo_train, sig_train), axis = 0)
                X_pseudo_test = np.concatenate((X_pseudo_test, sig_test), axis = 0)
```

```python
output = './processed_data/'
data_type = 'XDAWN_extended'
np.save(output + f'X_pseudo_test_{data_type}.npy', X_pseudo_test)
np.save(output + f'X_word_test_{data_type}.npy', X_word_test)
np.save(output + f'X_pseudo_train_{data_type}.npy', X_pseudo_train)
np.save(output + f'X_word_train_{data_type}.npy', X_word_train)
np.save(output + f'X_pseudo_valid_{data_type}.npy', X_pseudo_valid)
np.save(output + f'X_word_valid_{data_type}.npy', X_word_valid)
```
