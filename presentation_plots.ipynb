```python
import mne
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
raw = mne.io.read_epochs_eeglab('./data/B01_exp1.set')
for word in range(0,4):
    ch = 0
    plt.figure(figsize=[5,5])
    plt.plot(raw._times_readonly, raw.get_data()[word][ch])
    plt.title(f'WORD \n Channel: {raw.ch_names[word]}, Word: {raw.event_tagtype[word]["word"]}')
    plt.ylabel('Voltage')
    plt.xlabel('Time [ms]')
plt.figure(figsize=[5,5])
plt.plot(raw._times_readonly, np.mean(raw.get_data()[:300][ch], axis=0))
plt.ylabel('Voltage')
plt.xlabel('Time [ms]')
```
